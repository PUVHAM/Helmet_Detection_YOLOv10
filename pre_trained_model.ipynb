```python
# Initializing the model
from ultralytics import YOLOv10

MODEL_PATH = 'yolov10n.pt'
model = YOLOv10(MODEL_PATH)
```

```python
# Download images to predict
!gdown '1tr9PSRRdlC2pNir7jsYugpSMG-7v32VJ' -O './images/'
```

```python
# Predict the image
IMG_PATH = './images/HCMC_Street.jpg'
result = model(source=IMG_PATH)[0]
```

```python
# Save prediction result
result.save('./images/HCMC_Street_predict.jpg')
```

```python
!pip install pytube

import pytube
import cv2

# Predict a youtube video
YOUTUBE_VIDEO_PATH = 'https://youtu.be/wqPSsu7XQ74'

# Download the video using pytube
youtube = pytube.YouTube(YOUTUBE_VIDEO_PATH)
video = youtube.streams.get_highest_resolution()
video.download(filename='video.mp4') # Downloads into current working directory

video_results = model(source='video.mp4') # Change source to local video file

# Iterate through results and save each one
fourcc = cv2.VideoWriter_fourcc(*'mp4v') # Define video codec
out = cv2.VideoWriter('video_prediction.mp4', fourcc, 20.0, (video_results[0].orig_shape[1], video_results[0].orig_shape[0])) # Create VideoWriter object

for result in video_results:
    frame = result.plot() # Plot the result
    out.write(frame) # Write the frame to the video file

out.release() # Release the VideoWriter object
```
